## Visualization

##### Check for Outliers & Data Distribution

```python
# Boxplots to check outliers
numeric_cols = ['rating', 'marked_price', 'discounted_price', 'discount_amount', 'discount_percent']
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numeric_cols])
plt.xticks(rotation=45)
plt.title("Boxplot of Numeric Columns")
plt.show()
```

###### 1️⃣ Presence of Outliers  
- There are significant outliers in `marked_price`, `discounted_price`, and `discount_amount`.  
- This indicates that some products have **extremely high prices** or **massive discounts** compared to the majority.  

###### 2️⃣ Distribution of Prices  
- The **median marked price** is relatively low, meaning most products are affordable.  
- However, a **few expensive products (luxury items)** push the maximum price upwards.  

###### 3️⃣ Discount Trends  
- `discount_percent` seems to have a **narrower range**, meaning most discounts are within a **standard range**.  
- `discount_amount`, on the other hand, has **many outliers**, indicating some products have unusually high discounts.  

###### 4️⃣ Rating Analysis  
- The `rating` column does not show significant outliers, suggesting **most products have ratings within a reasonable range**.  

###### 🔍 Next Steps  
- **Investigate Outliers:** Identify which products have extremely high prices and discounts.  
- **Check Product Categories:** Do specific categories (e.g., luxury brands) contribute to these outliers?  
- **Analyze Correlations:** Do higher discounts lead to better ratings?  


```python
# remove outliers using IQR
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()  # Copy of DataFrame
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)  # First quartile
        Q3 = df_clean[col].quantile(0.75)  # Third quartile
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean

# Selecting numeric columns
numeric_columns = ["rating", "marked_price", "discounted_price", "discount_amount", "discount_percent"]

# Create a copy of original data
df_cleaned = remove_outliers_iqr(df.copy(), numeric_columns)

# Boxplot after cleaning
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned[numeric_columns])
plt.title("Boxplot of Numeric Columns")
plt.show()
```

##### Boxplot Analysis (After Removing Outliers)

###### 1. **Rating Distribution**
- Ratings are now more tightly packed, indicating a more uniform distribution.
- No extreme rating values suggest that most products have similar customer feedback.
###### 2. **Marked Price Trends**
- The price distribution has become more concentrated, removing extreme high-end prices.
- The median marked price suggests a balanced price range for most products.
###### 3. **Discounted Price Observations**
- Discounts have a more controlled range, with fewer extreme discounts affecting the dataset.
- The interquartile range (IQR) is more defined, ensuring that the majority of products have reasonable discounts.
###### 4. **Discount Amount and Discount Percent**
- Most products fall within a more reasonable discount range after outlier removal.
- Unusually high discounts have been eliminated, making the discount insights more reliable.

3. **Negative Correlations**

```python
# Average price of products
avg_price = df['discounted_price'].mean()
print(f"Average Discounted Price: {avg_price:.2f}")

# Count of highly discounted products
high_discount = df[df['discount_percent'] > 50].shape[0]
print(f"Number of products with more than 50% discount: {high_discount}")

# 📊 Visualization
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Average Discounted Price Distribution
sns.histplot(df['discounted_price'], bins=30, kde=True, color="blue", ax=ax[0])
ax[0].axvline(avg_price, color='red', linestyle='dashed', linewidth=2, label=f'Avg Price: ₹{avg_price:.2f}')
ax[0].set_title("Distribution of Discounted Prices")
ax[0].set_xlabel("Discounted Price (₹)")
ax[0].set_ylabel("Count of Products")
ax[0].legend()

# Plot 2: Count of Highly Discounted Products
labels = ['> 50% Discount', '≤ 50% Discount']
sizes = [high_discount, len(df) - high_discount]
colors = ['red', 'blue']
ax[1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
ax[1].set_title("Proportion of Highly Discounted Products")

# Show plots
plt.tight_layout()
plt.show()
```

###### 1️⃣ Distribution of Discounted Prices
- The **majority of discounted prices** are concentrated on the lower end, indicating that most products fall within an affordable price range.
- The **red dashed line** represents the **average discounted price (₹1532.62)**, showing that most products are priced below this threshold.
- The distribution is **right-skewed**, suggesting that while some products have high prices, they are fewer in number.

###### 2️⃣ Proportion of Highly Discounted Products
- **34.4% of products** have a discount greater than 50%, making them highly discounted.
- The remaining **65.6% of products** have a discount of 50% or less, meaning that most discounts are moderate.
- This indicates that a significant portion of products are offered at substantial discounts, possibly as part of promotions or clearance sales.

###### 🔍 Key Insights:
- The **low average discounted price** suggests that most products are in an affordable range.
- A **significant proportion of products** receive high discounts (>50%), highlighting competitive pricing strategies.
- The **right-skewed price distribution** indicates that while high-priced items exist, they are relatively rare in the dataset.

- **Home decor items** such as **floor lamps, ceiling lamps, and photo frames** also appear in the top 10, indicating that lighting and decorative products receive significant discounts.  
- The **average discount across the top 10 categories remains above 60%**, highlighting major markdowns on these product types.
